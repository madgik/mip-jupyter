# Federated Linear Regression -> sklearn model

Run linear regression through Platform Backend and use the returned sklearn model for prediction + serialization.


```python
import numpy as np
import pandas as pd
import joblib
from platform_backend_client import configure, FederatedLinearRegression

configure()
```

```python
payload = {
    "name": "Notebook linear regression",
    "data_model": "dementia:0.1",
    "datasets": ["edsd"],
    "x": ["age_value", "education_level"],
    "y": ["rightocpoccipitalpole"],
    "parameters": {},
    "filters": None,
    "preprocessing": {},
    "mip_version": "9.0.0",
}
payload
```

```python
model = FederatedLinearRegression(payload)

print("experiment uuid:", model.experiment_uuid)
print("status:", model.status)
print("coef_.shape:", np.asarray(model.coef_).shape)
print("intercept_:", model.intercept_)
```

```python
x_new = np.zeros((3, model.n_features_in_), dtype=float)
if model.n_features_in_ >= 1:
    x_new[1, 0] = 1.0
if model.n_features_in_ >= 2:
    x_new[2, 1] = 1.0

feature_names = getattr(model, "feature_names_in_", [f"x{i}" for i in range(model.n_features_in_)])
x_new_df = pd.DataFrame(x_new, columns=list(feature_names))

print("predict:", model.predict(x_new_df))
```

```python
model_path = "linear_regression_from_platform.joblib"
model.dump(model_path)
reloaded_model = joblib.load(model_path)
print("saved and reloaded model from:", model_path)
print("reloaded coef_.shape:", np.asarray(reloaded_model.coef_).shape)
```
